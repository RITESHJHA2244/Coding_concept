```python
# priority queue concept  using list method
class Priorityqueue:
    def __init__(self):
        self.item=[]

    def push(self,data,priority):
        index=0
        while index<len(self.item) and self.item[index][1]<=priority:
            index+=1
        self.item.insert(index,(data,priority))

    def is_empty(self):
        return len(self.item)==0 

    def pop(self):
        if  self.is_empty():
            print("your priority is empty")
        return self.item.pop(0)[0]

    def size(self):
        return len(self.item)
    
p=Priorityqueue()
p.push("ritesh",4)
p.push("arjun",7)
p.push("aasima",2)
p.push("agra",5)
p.push("aambika",1)

while not p.is_empty():
    print(p.pop())
```

```python
# Priority Queue  concept using linked list 
class Node:
    def __init__(self,item=None,priority=None,next=None):
        self.item=item
        self.priority=priority
        self.next=next

class Priorityqueue:
    def __init__(self):
        self.start=None
        self.item_count=0

    def push(self,data,priority):
        n=Node(data,priority)
        if  not self.start or priority< self.start.priority:
            n.next=self.start
            self.start=n 
        else:
            temp=self.start
            while temp.next!=None and temp.next.priority<=priority:
                temp=temp.next
            n.next=temp.next
            temp.next=n
        self.item_count+=1

    def is_empty(self):
        return self.start==None
    
    def pop(self):
        if self.is_empty():
            print("your priority is empty")
        data=self.start.item
        self.start=self.start.next
        self.item_count-=1
        return data

    def size(self):
        return self.item_count
    

p=Priorityqueue()
p.push("ritesh",4)
p.push("arjun",7)
p.push("aasima",2)
p.push("agra",5)
p.push("aambika",1)
while not p.is_empty():
    print(p.pop())
```
