```python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Circular Singly Linked List class
class CircularSinglyLinkedList:
    def __init__(self):
        self.start = None  # शुरू में लिस्ट खाली होती है

    # लिस्ट को प्रिंट करना
    def print_list(self):
        if self.start is None:
            print("List is empty")
            return
        temp = self.start
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.start:
                break
        print("(back to start)")

    # लिस्ट की शुरुआत में जोड़ना
    def insert_at_start(self, data):
        new_node = Node(data)
        if self.start is None:
            self.start = new_node
            new_node.next = new_node
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.start
            self.start = new_node

    # लिस्ट के अंत में जोड़ना
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.start is None:
            self.start = new_node
            new_node.next = new_node
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.start

    # किसी data को सर्च करना
    def search(self, data):
        if self.start is None:
            return None
        temp = self.start
        while True:
            if temp.data == data:
                return temp
            temp = temp.next
            if temp == self.start:
                break
        return None

    # किसी node के बाद नया node जोड़ना
    def insert_after(self, prev_node, data):
        if prev_node is None:
            print("Given node not found!")
            return
        new_node = Node(data)
        new_node.next = prev_node.next
        prev_node.next = new_node

    # किसी node को delete करना
    def delete(self, data):
        if self.start is None:
            return

        temp = self.start

        # अगर सिर्फ एक node है
        if temp.data == data and temp.next == self.start:
            self.start = None
            return

        # अगर पहला node delete करना हो
        if temp.data == data:
            last = self.start
            while last.next != self.start:
                last = last.next
            self.start = temp.next
            last.next = self.start
            return

        # अन्य किसी node को delete करना
        prev = None
        while temp.next != self.start:
            if temp.data == data:
                break
            prev = temp
            temp = temp.next

        if temp.data == data:
            prev.next = temp.next

# --- लिस्ट का उपयोग ---
cll = CircularSinglyLinkedList()
cll.insert_at_start(20)
cll.insert_at_start(10)
cll.insert_at_end(30)
cll.insert_after(cll.search(20), 25)
cll.print_list()

cll.delete(30)
cll.insert_after(cll.search(25), 40)
cll.print_list()
```
