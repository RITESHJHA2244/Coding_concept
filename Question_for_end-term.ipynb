```python
# maximum depth of binary tree
class Node:
    def __init__(self, item=None, left=None, right=None):
        self.item = item
        self.left = left
        self.right = right

def max_depth(root):
    if root is None:
        return 0
    left_depth = max_depth(root.left)
    right_depth = max_depth(root.right)
    return 1 + max(left_depth, right_depth)

# Example binary tree:
#         1
#        / \\
#       2   3
#      / 
#     4

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)

print("Maximum Depth of the Binary Tree is:", max_depth(root))
```

```python
# find height of a binary tree
class Node:
    def __init__(self, item=None, left=None, right=None):
        self.item = item
        self.left = left
        self.right = right

def height(root):
    if root is None:
        return -1  # height of empty tree is -1
    left_height = height(root.left)
    right_height = height(root.right)
    return 1 + max(left_height, right_height)

# Example binary tree:
#         1
#        / \
#       2   3
#      / 
#     4

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)

print("Height of the Binary Tree is:", height(root))
```

```python
# check for same binary tree
class Node:
    def __init__(self, item=None, left=None, right=None):
        self.item = item
        self.left = left
        self.right = right

def is_same_tree(p, q):
    if p is None and q is None:
        return True
    if p is None or q is None:
        return False
    if p.item != q.item:
        return False
    return is_same_tree(p.left, q.left) and is_same_tree(p.right, q.right)

# Example:
# Tree 1
root1 = Node(1)
root1.left = Node(2)
root1.right = Node(3)

# Tree 2
root2 = Node(1)
root2.left = Node(2)
root2.right = Node(3)

print("Are the trees exactly the same?", is_same_tree(root1, root2))
```

```python
# binary tree level order traversal
class Node:
    def __init__(self, item=None, left=None, right=None):
        self.item = item
        self.left = left
        self.right = right

def level_order(root):
    if root is None:
        return []
    
    result = []
    queue = [root]  # Queue to store nodes for level-order traversal
    
    while queue:  
        node = queue.pop(0)  # Get the first node in the queue
        result.append(node.item)  # Add node value to result
        
        if node.left:
            queue.append(node.left)  # Add left child to queue
        if node.right:
            queue.append(node.right)  # Add right child to queue
    
    return result

# Example Binary Tree:
#         1
#        / \
#       2   3
#      / \
#     4   5

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print("Level Order Traversal:", level_order(root))
```

```python
# check if a tree is a binary search tree

class Node:
    def __init__(self, item=None, left=None, right=None):
        self.item = item
        self.left = left
        self.right = right

def is_bst(root):
    def inorder(node):
        nonlocal prev
        if node is None:
            return True
        if not inorder(node.left):
            return False
        if prev is not None and node.item <= prev:
            return False
        prev = node.item
        return inorder(node.right)

    prev = None
    return inorder(root)

# Example usage:

# BST Example:
#       10
#      /  \
#     5   15
#        /
#       12

root = Node(10)
root.left = Node(5)
root.right = Node(15)
root.right.left = Node(12)

print("Is this a Binary Search Tree?", is_bst(root))  # Output: True
```

```python
# find kth smallest term in binary tree
class Node:
    def __init__(self, item=None, left=None, right=None):
        self.item = item
        self.left = left
        self.right = right

class BST:
    def __init__(self):
        self.root = None

    def insert(self, data):
        self.root = self._insert(self.root, data)

    def _insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.item:
            root.left = self._insert(root.left, data)
        elif data > root.item:
            root.right = self._insert(root.right, data)
        return root

    def kth_smallest(self, root, k):
        # In-order traversal to get elements in sorted order
        result = []
        self._inorder_traversal(root, result)
        if k <= len(result):
            return result[k - 1]
        return None

    def _inorder_traversal(self, root, result):
        if root:
            self._inorder_traversal(root.left, result)
            result.append(root.item)
            self._inorder_traversal(root.right, result)

# Example usage:
# Creating BST
#        5
#       / \
#      3   7
#     / \   \
#    2   4   8

bst = BST()
bst.insert(5)
bst.insert(3)
bst.insert(7)
bst.insert(2)
bst.insert(4)
bst.insert(8)

k = 3
print(f"The {k}-th smallest element is:", bst.kth_smallest(bst.root, k))  # Output: 4
```

```python
# Find lowest common investor of two nodes in a BST
class Node:
    def __init__(self, item=None, left=None, right=None):
        self.item = item
        self.left = left
        self.right = right

class BST:
    def __init__(self):
        self.root = None

    def insert(self, data):
        self.root = self._insert(self.root, data)

    def _insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.item:
            root.left = self._insert(root.left, data)
        elif data > root.item:
            root.right = self._insert(root.right, data)
        return root

    def lowest_common_ancestor(self, root, n1, n2):
        # If the root is None, there's no LCA
        if root is None:
            return None
        
        # If both n1 and n2 are smaller than root, LCA lies in left subtree
        if n1 < root.item and n2 < root.item:
            return self.lowest_common_ancestor(root.left, n1, n2)
        
        # If both n1 and n2 are greater than root, LCA lies in right subtree
        if n1 > root.item and n2 > root.item:
            return self.lowest_common_ancestor(root.right, n1, n2)
        
        # If one of n1 or n2 is smaller and the other is greater than root,
        # then root is the LCA
        return root

# Example usage:
# Creating BST
#        5
#       / \
#      3   7
#     / \   \
#    2   4   8

bst = BST()
bst.insert(5)
bst.insert(3)
bst.insert(7)
bst.insert(2)
bst.insert(4)
bst.insert(8)

n1 = 2
n2 = 4
lca = bst.lowest_common_ancestor(bst.root, n1, n2)
print(f"The Lowest Common Ancestor of {n1} and {n2} is:", lca.item)  # Output: 3
```

```python
# find middle.py
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        
    def insertNode(self, data):
        nd = Node(data)
        if not self.head:
            self.head = nd
            return
        currNode = self.head
        while currNode.next:
            currNode = currNode.next
        currNode.next = nd

    def display(self):
        currNode = self.head
        while currNode:
            print(currNode.data, end=" -> ")
            currNode = currNode.next
        print("None")

    def showMiddle(self):
        slow = self.head
        fast = self.head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        if slow:
            print("Middle element:", slow.data)
        else:
            print("The list is empty.")


ll = LinkedList()


arr = [9, 3, 2, 4, 5, 7, 8, 10]
for item in arr:
    ll.insertNode(item)

print("Linked List:")
ll.display()

ll.showMiddle()
```

```python
# nextgreater.py
def nextGreaterElement(nums):
    stack = []
    result = [-1] * len(nums)

    for i in range(len(nums)):
        while stack and nums[stack[-1]] < nums[i]:
            index = stack.pop()
            result[index] = nums[i]
        
        
        stack.append(i)

    return result


nums = [6, 9, 1, 12, 4]
print("Next Greater Element:", nextGreaterElement(nums))
```

```python
# webBrowser.py
class Browser:
    def __init__(self, homepage):
        self.back_stack = []
        self.forward_stack = []
        self.current = homepage

    def visit(self, url):
        self.back_stack.append(self.current) 
        self.current = url
        self.forward_stack.clear()

    def back(self):
        if self.back_stack: 
            self.forward_stack.append(self.current)
            self.current = self.back_stack.pop()

    def forward(self):
        if self.forward_stack:
            self.back_stack.append(self.current)
            self.current = self.forward_stack.pop()

    def get_page(self):
        return self.current

browser = Browser("google.com")
browser.visit("github.com")
browser.visit("stackoverflow.com")
browser.back()
print(browser.get_page())
browser.back()
print(browser.get_page())
browser.back()
print(browser.get_page())
```

```python
# ✅ Code: Find Depth of a Node in a Binary Tree
class Node:
    def __init__(self, item):
        self.item = item
        self.left = None
        self.right = None

def find_depth(root, target, depth=0):
    if root is None:
        return -1
    if root.item == target:
        return depth

    # Search in left subtree
    left_depth = find_depth(root.left, target, depth + 1)
    if left_depth != -1:
        return left_depth

    # Search in right subtree
    return find_depth(root.right, target, depth + 1)

# Example Binary Tree:
#         1
#        / \
#       2   3
#      / \
#     4   5

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

target = 3
depth = find_depth(root, target)
if depth != -1:
    print(f"Depth of node {target} is: {depth}")
else:
    print("Node not found.")

```

```python
#1. reverse a string using stack

def reverse_string_using_stack(s):
    # Create an empty stack
    stack = []

    # Push all characters of the string into the stack
    for char in s:
        stack.append(char)

    # Pop characters from the stack and build the reversed string
    reversed_str = ''
    while stack:
        reversed_str += stack.pop()

    return reversed_str

# Example usage
input_str = "hello"
print("Original String:", input_str)
print("Reversed String:", reverse_string_using_stack(input_str))

```

```python
#11. sorting in link list
class Node:
    def __init__(self, item=None, next=None):
        self.item = item
        self.next = next

class Sll:
    def __init__(self, start=None):
        self.start = start

    def is_empty(self):
        return self.start == None

    def insert_at_start(self, data):
        n = Node(data, self.start)
        self.start = n

    def insert_at_last(self, data):
        n = Node(data)
        if not self.is_empty():
            temp = self.start
            while temp.next is not None:
                temp = temp.next
            temp.next = n
        else:
            self.start = n

    def search(self, data):
        temp = self.start
        while temp is not None:
            if temp.item == data:
                return temp
            temp = temp.next
        return None

    def insert_after(self, temp, data):
        if temp is not None:
            n = Node(data, temp.next)
            temp.next = n

    def print_list(self):
        temp = self.start
        while temp is not None:
            print(temp.item, end=' -> ')
            temp = temp.next
        print('None')

    def delete_last(self):
        if self.start is None:
            pass
        elif self.start.next is None:
            self.start = None
        else:
            temp = self.start
            while temp.next.next is not None:
                temp = temp.next
            temp.next = None

    def delete_item(self, data):
        if self.start is None:
            pass
        elif self.start.next is None:
            if self.start.item == data:
                self.start = None
        else:
            temp = self.start
            if temp.item == data:
                self.start = temp.next
            else:
                while temp.next is not None:
                    if temp.next.item == data:
                        temp.next = temp.next.next
                        break
                    temp = temp.next

    # ✅ Bubble sort for singly linked list
    def bubble_sort(self):
        end = None
        while end != self.start:
            p = self.start
            while p.next != end:
                q = p.next
                if p.item > q.item:
                    # Swap values
                    p.item, q.item = q.item, p.item
                p = p.next
            end = p

# ✅ Testing the modified list with bubble sort
mylist = Sll()
mylist.insert_at_start(20)
mylist.insert_at_start(10)
mylist.insert_at_last(30)
mylist.insert_after(mylist.search(20), 25)
mylist.print_list()

mylist.delete_item(30)
print()
mylist.insert_after(mylist.search(25), 40)
mylist.print_list()

# Call bubble sort
print("\nAfter Sorting:")
mylist.bubble_sort()
mylist.print_list()
```

```python
#5. find largest element in array
def find_largest(arr):
    if not arr:
        return None  # Handle empty array
    largest = arr[0]
    for num in arr[1:]:
        if num > largest:
            largest = num
    return largest

# Example usage
arr = [5, 3, 9, 1, 12, 7]
print("Largest element:", find_largest(arr))
```
